# Movie Recomendation Using CountVectorizer from sklearn 


```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
fx= pd.read_csv("/home/varunjaggi/Desktop/movie_dataset.csv")
```

```python
fx.head()
```

```python
features = ['keywords','cast','genres','director'] #To combine each row's required factors & return 
def combinefea(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+row['director']
```

```python
for feature in features: #to fill blanks instead of NaNs
    fx[feature] = fx[feature].fillna(' ')
fx.head() #after filling blanks
```

```python
fx["comb"]=fx.apply(combinefea,axis=1)
cv=CountVectorizer()
count_matrix=cv.fit_transform(fx["comb"]) #finding the simillarity and plotting them as vectors
count_matrix
```

```python
cosine_matrix=cosine_similarity(count_matrix) 
cosine_matrix
```

```python
def get_title_from_index(index): #to fetch title from index of the movie
    return fx[fx.index == index]["title"].values[0]
def get_index_from_title(title):# to fetch index from the title of the movie
    return fx[fx.title == title]["index"].values[0]
```

```python
def find_simillar(movie_user_likes):
    movie_index = get_index_from_title(movie_user_likes)
    similar_movies = list(enumerate(cosine_matrix[movie_index]))

    sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
    print("Top 5 similar movies to "+movie_user_likes+" are:\n")
    i=0
    for element in sorted_similar_movies:
        print(get_title_from_index(element[0]))
        i=i+1
        if i>5:
            break
    
```

```python
find_simillar("Skyfall")
```
